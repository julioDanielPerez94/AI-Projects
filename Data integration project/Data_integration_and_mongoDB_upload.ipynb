5. Diferentes valores para la misma información

### 4. Definimos nuestras funciones, que nos servirán para tratar nuestro conjunto de datos, que posteriormente cargaremos en MongoDB

```python
def age(birthdate):
    '''
    Esta función realiza lo siguiente:
    1. Le pasamos una fecha, que sería la de nacimiento, aportado
    por uno de los campos
    2. Calcula el tiempo que ha pasado hasta hoy
    3. Nos devuelve la edad que tendría la persona
    '''
    birthdate = pd.to_datetime(birthdate)
    today = date.today()
    age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
    return age

def convert_to_a_list(lista):
    '''
    Esta función nace con la idea de homogeneizar el formato de 
    DNI_Amigos, que proviene de la fuente 1 de datos. Realiza lo siguiente::
    1. Le pasamos cada valor de DNI_Amigos, que es una lista en formato
    [{'DNI':XXXX ,'DNI':XXXX}]
    2. Cogemos los valores XXXX y los añadimos a un vector vacío m
    3. Convertimos cada valor en entero
    4. Nos devuelve la lista con los valores de DNI y en formato int()
    '''
    m = []
    for x in lista:
    # print(x['DNI'])

        m = np.append(m, int(x['DNI']))
  
    m = list(m)
    for j,i in enumerate(m):
        m[j] = int(i)

    return m

def fixing_columns (df_test_1):
    df_test = df_test_1.copy()
    '''
    En esta función agruparemos todas las acciones que
    llevaremos a cabo para poder transformar, limpiar y 
    dejar listo el dataframe, que posteriormente cargaremos
    en la BBDD
    '''
    # Aplicamos la función convert_to_a_list al campo de los DNI_Amigos(Aficiones_A) de la fuente 1
    df_test.iloc[:, 7] = df_test.iloc[:,7].apply(lambda x: convert_to_a_list(x) if type(x) == list else x)
    for i in range(len(df_test)):
        ########################################
        # Arreglando la columna de Nombres y Apellidos
        ########################################
        '''
        Para poder arreglar las columnas de nombres y apellidos, hacemos lo siguiente:
        1. Comprobamos si los campos Nombres y Apellidos de la fuente 2 son NaN. Si lo 
        son, insertamos los datos de la fuente 1
        2. Comprobamos si los campos de Nombres y Apellidos de la fuente 1 son NaN. si lo 
        son, insertamos los datos de la fuente2
        3. Comprobamos si los campos de Nombres y Apellidos de la fuente 1 y 2 son string,
        que implica que tienen datos. Si lo son, comprobamos que tengan el mismo valor y si lo tiene 
        no hacemos nada. Si no, integraríamos en cada campo el valor combinado de ambas fuente.
        Como vemos en el resultado final, esto no pasa.
        '''
        if (type(df_test.iloc[i,1]) == str) & (type(df_test.iloc[i,2]) == str) & (type(df_test.iloc[i,9]) == float) & (type(df_test.iloc[i,8]) == float):
            df_test.iloc[i,9] = df_test.iloc[i,1]
            df_test.iloc[i,8] = df_test.iloc[i,2]
        elif (type(df_test.iloc[i,1]) != str) & (type(df_test.iloc[i,2]) != str) & (type(df_test.iloc[i,9]) == str) & (type(df_test.iloc[i,8]) == str):
            df_test.iloc[i,1] = df_test.iloc[i,9]
            df_test.iloc[i,2] = df_test.iloc[i,8]
        elif (type(df_test.iloc[i,1]) == str) & (type(df_test.iloc[i,2]) == str) & (type(df_test.iloc[i,9]) == str) & (type(df_test.iloc[i,8]) == str):
            if (df_test.iloc[i,1] == df_test.iloc[i,9]) & (df_test.iloc[i,2] == df_test.iloc[i,8]):
                pass
        
            else:
                a = df_test.iloc[i,1]
                b = df_test.iloc[i,2]
                c = df_test.iloc[i,9]
                d = df_test.iloc[i,8]
                df_test.iloc[i,1] = a + '/' + c
                df_test.iloc[i,9] = a + '/' + c
                df_test.iloc[i,2] = b + '/' + d
                df_test.iloc[i,8] = b + '/' + d
        ########################################
        # Arreglando la columna de Ciudad
        ########################################

        '''
        Para poder actuar sobre los campos de Ciudad, hacemos lo siguiente:
        1. Comprobamos si el campo de Ciudad de la fuente 2 es NaN. Si lo 
        es, insertamos los datos de la fuente 1
        2. Comprobamos si el campos de Ciudad de la fuente 2 es NaN. Si lo 
        es, insertamos los datos de la fuente 2.
        3. Comprombamos si el campo de Ciudad de la fuente 1 y 2 son string,
        que implica que tienen datos. Si lo son, comprobamos que tengan el mismo valor y si lo tiene 
        no hacemos nada. Si no, integraríamos en cada campo el valor combinado de ambas fuente.
        En este caso, si hay una inconsistencia en los datos, por lo que se combinan
        y se actualizan los valores, para indicar que ahí si hay un 
        problema con ese valor.        
        '''
        if (type(df_test.iloc[i,3]) == str) & (type(df_test.iloc[i,10]) == float):
            df_test.iloc[i,10] = df_test.iloc[i,3]
        elif (type(df_test.iloc[i,3]) == float) & (type(df_test.iloc[i,10]) == str) :
            df_test.iloc[i,3] = df_test.iloc[i,10]
        elif (type(df_test.iloc[i,3]) == str) & (type(df_test.iloc[i,10]) == str):
            if df_test.iloc[i,3] == df_test.iloc[i,10]:
                pass
            else: 
                a = df_test.iloc[i,3]
                b = df_test.iloc[i,10]
                df_test.iloc[i,3] = a + '/' + b
                df_test.iloc[i,10] = a + '/' + b
        else:
            pass


        ########################################
        # Arreglando las columnas de Sexo
        ########################################
        '''
        En este campo exite una incoherencia en los valores,
        ya que ambos campos indican la misma información de diferente
        forma. Por tanto, para homogeneizar el campo de sexo haremos lo siguiente:
        1. Comprobamos si el campo Sexo de la fuente 1 es NaN y, si lo es
        y además el valor de Sexo de la fuente 2 es 0, le asignamos
        al primero y segundo el valor M
        2. Comprobamos si el campo Sexo de la fuente 1 es NaN y, si no lo es
        y además el valor de Sexo de la fuente 2 es 0, le asignamos
        al segundo el valor M
        3. Comprobamos si el campo Sexo de la fuente 1 es NaN y, si lo es
        y además el valor de Sexo de la fuente 2 es 1, le asignamos
        al primero y segundo el valor H
        4. Comprobamos si el campo Sexo de la fuente 1 es NaN y, si no lo es
        y además el valor de Sexo de la fuente 2 es 1, le asignamos
        al segundo el valor H
        5. Comprobamos si el campo Sexo de la fuente 2 es NaN, y si 
        lo fuera y además el valor de Sexo de la fuente 2 no es NaN,
        el primero tomará el valor del segundo
        De esta manera homogeneizamos los valores de Mujer y Hombre
        a M y H en ambos campos
        '''
        
        if (type(df_test.iloc[i,5]) != str) & (df_test.iloc[i,12] == 0):
            df_test.iloc[i,5] = 'M'
            df_test.iloc[i,12] = 'M'
        elif (type(df_test.iloc[i,5]) == str) & (df_test.iloc[i,12] == 0):
            df_test.iloc[i,12] = 'M'
        elif (type(df_test.iloc[i,5]) != str) & (df_test.iloc[i,12] == 1):
            df_test.iloc[i,5] = 'H'
            df_test.iloc[i,12] = 'H'
        elif (type(df_test.iloc[i,5]) == str) & (df_test.iloc[i,12] == 1):
            df_test.iloc[i,12] = 'H'
        elif np.isnan(df_test.iloc[i,12]) & (type(df_test.iloc[i,5]) == str):
            df_test.iloc[i,12] = df_test.iloc[i,5]
        else: 
            pass

        ########################################
        # Arreglando las columna de Edad
        ########################################
        '''
        Para arreglar el campo Edad, de la fuente 1,
        optamos por aplicar la función age() pasándole 
        las fechas de nacimiento de cada personas.
        '''
        if np.isnan(df_test.iloc[i,4]) == True:
            df_test.iloc[i,4] = age(df_test.iloc[i, 11])

        ########################################
        # Arreglando los NaN de Fecha_nacimiento y Estatura
        ########################################
        '''
        Para arreglar las columnas de Fecha de nacimiento y estatura
        optamos por indicar '01-01-1900' en aquellos valores en 
        donde no tenemos la fecha. Además, como es un campo exclusivo
        de la fuente 2 de datos, no hacemos más acción.
        En cuando a la estatura pasa lo mismo, por lo que nos 
        limitamos a insertar el valor 0 en los valores NaN.        
        '''

        if type(df_test.iloc[i,11]) != str:
          # df_test.iloc[i,11] = pd.to_datetime('01-01-1900', format = '%d-%m-%y')
          df_test.iloc[i,11] = '01-01-1900'
        if np.isnan(df_test.iloc[i,13]) == True:
            df_test.iloc[i,13] = 0.0

        ########################################
        # Arreglando los DNI de amigos
        ########################################
        '''
        Para arreglar los DNI de los amigos, habíamos optado
        por igualar por:
        1. Insertar donde hubiera NaN el valor del campo de la fuente 
        opuesta
        2. En el caso de haber valores en ambos campos, comprobar si 
        son iguales. En caso de no serlo, optar por el que tiene
        más información
        
        Finalmente, hemos decidido incluir N/D en aquellos valores
        de estos campos que fueran Nan, para poder cumplir con 
        uno de los requisitos, que era poder distinguir las fuentes
        de datos, ya que sino, lo que obtendríamos serían dos 
        campos totalmente iguales, pero con distino nombre.
        '''
        if (type(df_test.iloc[i, 7]) == float):
            df_test.iloc[i, 7] = ['N/D']
        if (type(df_test.iloc[i, 14]) == float):
            df_test.iloc[i, 14] = ['N/D']
        
        # para la versión de Google Colaboratory, habíamos aplicado los cambios
        # comentados en la decripción
#        p = [] # Definimos una lista vacía
#         if (type(df_test.iloc[i, 7]) == float) & (type(df_test.iloc[i, 14]) == list):
# #             df_test.iloc[i, 7] = list(np.append(int(i), p) for i in df_test.iloc[i, 14])
#             df_test.iloc[i, 7] = df_test.iloc[i, 14]
#         # elif (type(df_test.iloc[i, 7]) == list) & (type(df_test.iloc[i, 14]) == list):
#         #   print('opcion 2')
#         #   df_test.iloc[i, 7] = .apply(lambda x: convert_to_a_list(x) if type(x) == list else x)
#         elif (type(df_test.iloc[i, 7]) == list) & (type(df_test.iloc[i, 14]) == float):
# #             df_test.iloc[i, 14] =  list(np.append(int(i), p) for i in df_test.iloc[i, 7])
#             df_test.iloc[i, 14] = df_test.iloc[i, 7]
#         elif (type(df_test.iloc[i, 7]) == float) & (type(df_test.iloc[i, 14]) == float):
#             df_test.iloc[i, 7] = ['N/D']
#             df_test.iloc[i, 14] = ['N/D']
#         else:
#             pass

        
            
        ########################################
        # Arreglando las aficiones
        ########################################
        '''
        Para arreglar los campos de las Aficiones,
        hemos tomado el mismo criterio que para los DNIs
        de los amigos, es decir, sustituyendo los valores
        nulos por N/D
        '''

        if (type(df_test.iloc[i, 6]) == list) & (type(df_test.iloc[i, 15]) == float):
            df_test.iloc[i, 15] = ['N/D']
        elif (type(df_test.iloc[i, 6]) != list) & (type(df_test.iloc[i, 15]) != float):
            df_test.iloc[i, 6] = ['N/D']
        # if (type(df_test.iloc[i, 6]) == float) & (type(df_test.iloc[i, 15]) == list):
        #   df_test.iloc[i, 6] = list(np.append(i, p) for i in df_test.iloc[i, 15])
        else:
            pass

    # cambiamos el formato del campo Fecha de nacimiento    
    df_test.iloc[:, 11] = pd.to_datetime(df_test.iloc[:, 11])
    '''
    Finalmente, le cambiamos el nomber a todos los campos,
    para poder identificar los campos que proceden de 
    la fuente 1 (_A) y la fuente 2(_B)
    '''
    df_test.columns=['DNI', 'Nombre_A', 'Apellidos_A', 'Ciudad_A', 'Edad_A', 'Sexo_A',
       'Aficiones_A', 'DNI_Amigos_A', 'Apellidos_B', 'Nombre_B', 'Ciudad_B',
       'Fecha_nacimiento_B', 'Sexo_B', 'Estatura_B', 'DNI_Amigos_B', 'Aficiones_B']

    return df_test # devolvemos el dataframe tratado
```

### 5. Aplicamos nuestra función al conjunto de datos anteriormente integrados y vemos el resulto. También exportamos el fichero al entorno local.

```python
df_fixed = fixing_columns(df_test) # aplicamos el 
df_fixed.to_csv('df_fixed.csv', index= False) # exportamos los datos
df_fixed
```

### 6. Cargamos los datos en una BBDD de MongoDB utilizando la librería Pymongo

```python
from pymongo import MongoClient # importamos el cliente Mongo
```

```python
# https://unix.stackexchange.com/questions/434791/how-to-create-a-new-database-and-collection-using-pymongo
# mongo es lazy, lo que supone que la base de datos se crea una vez metes los datos
# https://pythonexamples.org/python-mongodb-create-collection/

cliente = MongoClient('mongodb://localhost:27017/') # nos conectamos a la base de datos en localhost
```

```python
import csv # importamos la librería csv
mydb = cliente["mydatabase"] # creamos la base de datos mydatabase
mycol = mydb["personas"] # creamos nuestra colección customers

csvfile = open('df_fixed.csv', 'r', encoding='utf-8') # cargamos el fichero tratado que anteriormente hemos exportado
reader = csv.DictReader( csvfile ) # lo cargamos en la variable reader, usando el paquete csv
header = list(df_fixed.columns) # listamos las columnas que se van a transformar en los campos de nuestra BBDD

'''
Con este bucle que vemos a continuación, lo que hacemos es crear un conjunto de datos de estilo json 
con clave:valor, que se adapta perfectamente a lo que necesitamos para cargar los datos 
en MongoDB. Finalmente, vamos insertando cada row o conjunto de datos de uno en uno. 
'''
for each in reader:
    print('Imprimiendo each:', each, '\n')
    row={}
    for field in header:
        row[field]=each[field]
        
    print ('Imprimiendo row:',row, '\n')
    mycol.insert_one(row)

print(cliente.list_database_names()) # listamos los nombres de las bases de datos
print(mydb.list_collection_names()) # listamos el nombre de las colecciones creadas
```
